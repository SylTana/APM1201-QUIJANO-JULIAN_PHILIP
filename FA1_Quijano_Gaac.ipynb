```python
from itertools import permutations as perm

# R-1.1 Write a short Python function, is multiple(n, m), that takes two integer values and returns True if n is a multiple of 
# m, that is, n = mi for some integer i,and False otherwise.

def is_multiple(n, m):
    if (m % n) == 0:
        return True
    else:
        return False

x = int(input("X = "))
y = int(input("Y = "))

if (is_multiple(x, y)) == True:
    print("Is", x,"a multiple of", str(y)+"? \nYes")
else:
    print("Is", x,"a multiple of", str(y)+"? \nNo")

print("\n")

# P-1.29 Write a Python program that outputs all possible strings formed by using the characters c, a, t, d, o, and g 
# exactly once.

def permutation(a, b):
    c = list(perm(a, b))
    print(str(b) + "-letter combinations:")
    for i in c:
        print(''.join(i), end=', ')
    print("\n")
    print("There are a total of", len(c), str(b)+"-letter combinations.")
    print("\n")

def permutation_rec(a, b):
    while (b != 0):
        permutation(a, b)
        b -= 1

permutation_rec(["c", "a", "t", "d", "o", "g"], 6)
```
