### Perform experimental analysis to test the hypothesis that Python’s sorted method runs in O(nlogn) time on average.

```python
import random as r
import time as t
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd


# P-3.57 Perform experimental analysis to test the hypothesis that Python’s sorted method runs in O(nlogn) time on average.

data = []
a = []
time_data = []

b5000 = [i for i in range(1, 5001)]
for i in range(5000):
    a.append(r.randint(1, 100))
    r.shuffle(a)
    start_time = t.time()
    sorted(a)
    time_data.append(t.time_ns() - start_time)


t_df = pd.DataFrame(data, columns=['n', 'Time'])
t_df['n'] = b5000
t_df['Time'] = time_data

sb.scatterplot(data=t_df, x='Time', y='n', linewidth=0.1)
ax = plt.gca()
ax.grid(True)
ax.set_xscale("log")
ax.set_yscale("log")
plt.show()
```
